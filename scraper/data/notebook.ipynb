```python
import pandas as pd
import sqlite3

DB_PATH = "raw.db"

conn = sqlite3.connect(DB_PATH)
df = pd.read_sql_query("SELECT * FROM pages", conn)
conn.close()

df
```

```python

# Duplicados por contenido identico
duplicados = df[df.duplicated(subset="content", keep=False)].sort_values("content")

print(f"Filas con contenido duplicado: {len(duplicados)} de {len(df)} total")
print(f"Contenidos unicos duplicados: {duplicados['content'].nunique()}\n")

duplicados[["url", "title", "category", "content"]].assign(
    content_preview=duplicados["content"].str[:80]
).drop(columns="content")
```
